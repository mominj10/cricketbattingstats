```python
#to rename columns
df=df.rename(columns={'Mat':'Matches', 'NO':'Not_Outs', 'HS':'Highest_score','BF':'Balls_Faced','SR':'Strike_rate'})
```

```python
#check null values
df.isnull().any()
```

```python
#check whuch rows have null values
#df['Balls_Faced'].isna()==1 It creates a boolean mask with True for NaN values and False for non-NaN values in the 'Balls_Faced' column.
df[df['Balls_Faced'].isna()==1]
#df[df.isnull.any()]
```

```python
#to change null valuse
df['Balls_Faced']=df['Balls_Faced'].fillna(0)
df['Strike_rate']=df['Strike_rate'].fillna(0)
```

```python
#split a column SPAN into Start and END
df['Span'].str.split(pat='-')
```

```python
#select the new colummn name and = the column you wwant to change
#str[0] extracts the first part of the split string, representing the start year of the span.
df['Rookie_year']= df['Span'].str.split(pat='-').str[0]
```

```python
#str[1] extracts the second part of the split string, representing end year of the span.
df['final_year']= df['Span'].str.split(pat='-').str[1]
```
